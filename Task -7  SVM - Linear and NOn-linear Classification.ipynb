# 5. Splitting the Train and test data

```python
dataset.columns
```

```python
X = dataset[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
       'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
X
```

```python
Y =dataset[[ 'diagnosis_M']]
Y
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train / test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='rbf',random_state = 0)
clf.fit(X_train,y_train)
```

```python
y_predict = clf.predict(X_test)
y_predict
```

```python
from sklearn.svm import SVC
clf_1= SVC(kernel='linear',random_state = 0)
clf_1.fit(X_train,y_train)
```

# 6. Evaluation Metric

```python
from sklearn.metrics import confusion_matrix 
CM = confusion_matrix(y_test,y_predict)
CM
```

# 7. GridsearchCv -(Hyper Tuning)

```python
from sklearn.model_selection import GridSearchCV

params = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 0.1, 1, 10],
    'kernel': ['rbf','linear']
}

grid = GridSearchCV(SVC(), params, cv=5)
grid.fit(X_train, y_train)

print("Best parameters:", grid.best_params_)
print("Best cross-validation accuracy:", grid.best_score_)
```

```python
 # print best parameter after tuning
#print(grid.best_params_)
re=grid.cv_results_
print("The accuracy_score value for best parameter {}:".format(grid.best_params_))
```

# 8. Cross Validation

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(SVC(kernel='rbf', C=1, gamma='scale'),
                         X_train, y_train, cv=5)

print("Cross-validation scores:", scores)
print("Mean accuracy:", scores.mean())
```

# 9.Visualize decision boundary using PCA 2D data.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

# Load data
data = load_breast_cancer()
X = data.data
y = data.target

# Scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA reduce 30 → 2
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Train SVM on all 30 features (IMPORTANT)
model = SVC(kernel='linear')
model.fit(X_scaled, y)

# Predict on PCA grid
x_min, x_max = X_pca[:,0].min() - 1, X_pca[:,0].max() + 1
y_min, y_max = X_pca[:,1].min() - 1, X_pca[:,1].max() + 1

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 300),
    np.linspace(y_min, y_max, 300)
)

# Inverse PCA → original 30D → model.predict
grid_pca = np.c_[xx.ravel(), yy.ravel()]
grid_original = pca.inverse_transform(grid_pca)

Z = model.predict(grid_original)
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, edgecolor='k')
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("Decision Boundary (PCA 2D Projection of 30 Features)")
plt.show()
```
