# 1. Load Dataset

```python
# import the libraries
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("Breast_Cancer_data.csv")
dataset
```

# 2. Data Cleaning

```python
dataset.isnull().sum()
```

```python
# Drop the unwanted columns
dataset = dataset.drop(columns=['id'])
dataset = dataset.drop(columns=['Unnamed: 32'])
```

```python
dataset['diagnosis'].value_counts()
```

```python
dataset.isnull().sum()
```

# 3. Data Distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(5,3))
sns.countplot(x='diagnosis', data=dataset)   
plt.title('Distribution of Breast Cancer Classes')
plt.show()
```

# 4. Data Preprocessing

```python
dataset= pd.get_dummies(dataset,drop_first = True,)
dataset = dataset.replace({True:1,False:0})
dataset
```

# 5. Splitting the Train and test data

```python
dataset.columns
```

```python
X = dataset[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
       'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
print(X)
```

```python
Y = dataset[['diagnosis_M']]
print(Y)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train / test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, Y, test_size=0.2, random_state=42)
```

```python
# Training the Model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

# 6. Prediction and Evaluation

```python
y_pred = model.predict(X_test)

y_prob = model.predict_proba(X_test)[:, 1]
```

```python
# Confusion metrics
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Clf report
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
# ROC- AUC curve
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

roc_auc = roc_auc_score(y_test, y_prob)
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.plot(fpr, tpr, label=f'ROC Curve (AUC={roc_auc:.2f})')
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
